````python
# ============================================
# 📝 PROMPT TEMPLATES
# ============================================

from langchain_core.prompts import PromptTemplate

# ---- STEP 1: EXTRACTION PROMPT ----
EXTRACTION_TEMPLATE = """You are an expert HR analyst and resume parser.

CRITICAL RULES:
1. ONLY extract information EXPLICITLY mentioned in the resume.
2. Do NOT assume or hallucinate any skills not present.
3. If information is not available, use empty lists [] or null.
4. Be precise with years of experience.

Example:
Input: "John has 4 years as a data analyst. Proficient in Python, SQL, Tableau."
Output:
{{
    "candidate_name": "John",
    "technical_skills": ["Python", "SQL", "Tableau"],
    "years_of_experience": 4,
    "ml_frameworks": [],
    "cloud_platforms": []
}}

Now analyze this resume:

RESUME:
{resume}

Extract in this JSON format:
{{
    "candidate_name": "Full name",
    "technical_skills": ["all technical skills"],
    "soft_skills": ["soft skills mentioned"],
    "programming_languages": ["languages"],
    "ml_frameworks": ["ML/DL frameworks"],
    "cloud_platforms": ["cloud services"],
    "tools_and_frameworks": ["tools and libraries"],
    "years_of_experience": <total years as integer>,
    "experience_details": [
        {{
            "role": "title",
            "company": "company",
            "duration": "duration",
            "key_achievements": ["achievements"]
        }}
    ],
    "education": [
        {{
            "degree": "degree",
            "institution": "school",
            "year": "year"
        }}
    ],
    "certifications": ["certifications"],
    "leadership_experience": true/false,
    "publications": ["publications if any"]
}}

Return ONLY valid JSON."""

extraction_prompt = PromptTemplate(
    input_variables=["resume"],
    template=EXTRACTION_TEMPLATE
)

# ---- STEP 2: MATCHING PROMPT ----
MATCHING_TEMPLATE = """You are an expert recruitment analyst.

CRITICAL RULES:
1. ONLY match skills EXPLICITLY present in the candidate profile.
2. Do NOT assume skills not listed.
3. "basic Python" does NOT equal "expert Python".

CANDIDATE PROFILE:
{extracted_data}

JOB DESCRIPTION:
{job_description}

Compare and return JSON:
{{
    "required_skills_matched": [
        {{"skill": "skill name", "candidate_level": "expert/proficient/basic", "match_quality": "strong/partial/weak"}}
    ],
    "required_skills_missing": [
        {{"skill": "skill name", "importance": "critical/important"}}
    ],
    "preferred_qualifications_matched": [
        {{"qualification": "qual name", "match_quality": "strong/partial/weak"}}
    ],
    "preferred_qualifications_missing": [
        {{"qualification": "qual name"}}
    ],
    "experience_match": {{
        "required_years": 5,
        "candidate_years": <actual>,
        "meets_requirement": true/false
    }},
    "education_match": {{
        "meets_requirement": true/false,
        "details": "explanation"
    }},
    "overall_match_percentage": <0-100>,
    "match_summary": "2-3 sentence summary"
}}

Return ONLY valid JSON."""

matching_prompt = PromptTemplate(
    input_variables=["extracted_data", "job_description"],
    template=MATCHING_TEMPLATE
)

# ---- STEP 3: SCORING PROMPT ----
SCORING_TEMPLATE = """You are an expert recruitment scoring system.

SCORING RUBRIC (Total: 100):
- Technical Skills (0-30): Match quality of required technical skills
- Experience (0-25): Years and relevance of experience
- Education & Certifications (0-15): Degree level and relevance
- Tools & Technologies (0-15): Required tools proficiency
- Additional Value (0-15): Leadership, publications, preferred quals

MATCHING ANALYSIS:
{matching_result}

CANDIDATE PROFILE:
{extracted_data}

JOB DESCRIPTION:
{job_description}

Score the candidate:
{{
    "candidate_name": "name",
    "technical_skills_score": {{"score": <0-30>, "max_score": 30, "justification": "why"}},
    "experience_score": {{"score": <0-25>, "max_score": 25, "justification": "why"}},
    "education_score": {{"score": <0-15>, "max_score": 15, "justification": "why"}},
    "tools_technologies_score": {{"score": <0-15>, "max_score": 15, "justification": "why"}},
    "additional_value_score": {{"score": <0-15>, "max_score": 15, "justification": "why"}},
    "total_score": <0-100>,
    "candidate_tier": "Strong Fit / Moderate Fit / Weak Fit / Not Suitable",
    "recommendation": "Strongly Recommend / Recommend / Consider / Do Not Recommend"
}}

Return ONLY valid JSON."""

scoring_prompt = PromptTemplate(
    input_variables=["matching_result", "extracted_data", "job_description"],
    template=SCORING_TEMPLATE
)

# ---- STEP 4: EXPLANATION PROMPT ----
EXPLANATION_TEMPLATE = """You are an expert recruitment advisor.

Generate a clear, detailed explanation of the candidate's evaluation.

CANDIDATE PROFILE:
{extracted_data}

MATCHING ANALYSIS:
{matching_result}

SCORING BREAKDOWN:
{scoring_result}

JOB DESCRIPTION:
{job_description}

Provide evaluation report in JSON:
{{
    "candidate_name": "name",
    "overall_assessment": "3-4 sentence executive summary",
    "score": <total score>,
    "tier": "Strong Fit / Moderate Fit / Weak Fit / Not Suitable",
    "key_strengths": [
        {{"strength": "specific strength", "evidence": "evidence from resume"}}
    ],
    "key_gaps": [
        {{"gap": "specific gap", "impact": "how it affects candidacy"}}
    ],
    "detailed_reasoning": {{
        "why_this_score": "detailed explanation of score",
        "risk_factors": ["risks if hired"],
        "growth_potential": "growth assessment"
    }},
    "interview_recommendations": ["topics to explore in interview"],
    "final_recommendation": {{
        "decision": "Strongly Recommend / Recommend / Consider / Do Not Recommend",
        "confidence": "High / Medium / Low",
        "reasoning": "final reasoning"
    }}
}}

Return ONLY valid JSON."""

explanation_prompt = PromptTemplate(
    input_variables=["extracted_data", "matching_result", "scoring_result", "job_description"],
    template=EXPLANATION_TEMPLATE
)

print("✅ All 4 prompt templates created!")
print("   1. Extraction Prompt")
print("   2. Matching Prompt")
print("   3. Scoring Prompt")
print("   4. Explanation Prompt")
````

````python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.output_parsers import StrOutputParser

# Initialize FREE Gemini Model
llm = ChatGoogleGenerativeAI(
    model="gemini-2.5-flash-lite", # Changed from "gemini-2.5-pro" to "gemini-2.5-flash-lite"
    temperature=0.1,
    google_api_key=os.environ["GOOGLE_API_KEY"],
    convert_system_message_to_human=True,
)

# Create output parser
parser = StrOutputParser()

# ---- Build 4 Chains using LCEL (pipe operator) ----

# Chain 1: Resume → Extract Skills
extraction_chain = extraction_prompt | llm | parser

# Chain 2: Extracted Data + JD → Match
matching_chain = matching_prompt | llm | parser

# Chain 3: Match Results → Score
scoring_chain = scoring_prompt | llm | parser

# Chain 4: All Results → Explain
explanation_chain = explanation_prompt | llm | parser

print("✅ All 4 LangChain chains built using LCEL!")
print("   → extraction_chain = extraction_prompt | llm | parser")
print("   → matching_chain   = matching_prompt | llm | parser")
print("   → scoring_chain    = scoring_prompt | llm | parser")
print("   → explanation_chain = explanation_prompt | llm | parser")
````

````python
# ============================================
# 🔧 HELPER FUNCTIONS
# ============================================

import json
import re


def parse_json_safely(json_string):
    """Safely parse JSON from LLM output."""
    try:
        return json.loads(json_string)
    except json.JSONDecodeError:
        try:
            # Extract JSON from markdown code blocks
            match = re.search(r'```(?:json)?\s*([\s\S]*?)\s*```', json_string)
            if match:
                return json.loads(match.group(1))
            # Find JSON object in string
            match = re.search(r'\{[\s\S]*\}', json_string)
            if match:
                return json.loads(match.group(0))
        except:
            pass
    return {"error": "Failed to parse", "raw": json_string[:500]}


def print_report(results, candidate_label):
    """Print a formatted screening report."""

    scoring = parse_json_safely(results.get("scoring", "{}"))
    explanation = parse_json_safely(results.get("explanation", "{}"))

    print(f"\n{'='*70}")
    print(f"  📋 SCREENING REPORT: {candidate_label}")
    print(f"{'='*70}")

    name = explanation.get("candidate_name", scoring.get("candidate_name", "Unknown"))
    score = explanation.get("score", scoring.get("total_score", "N/A"))
    tier = explanation.get("tier", scoring.get("candidate_tier", "N/A"))

    print(f"\n  👤 Candidate: {name}")
    print(f"  📊 Score: {score}/100")
    print(f"  🏷️  Tier: {tier}")

    # Score Breakdown
    if "error" not in scoring:
        print(f"\n  📊 SCORE BREAKDOWN:")
        print(f"  {'-'*45}")
        for key, label in [
            ("technical_skills_score", "Technical Skills"),
            ("experience_score", "Experience"),
            ("education_score", "Education & Certs"),
            ("tools_technologies_score", "Tools & Tech"),
            ("additional_value_score", "Additional Value"),
        ]:
            cat = scoring.get(key, {})
            if isinstance(cat, dict):
                print(f"     {label:.<30} {cat.get('score','?')}/{cat.get('max_score','?')}")
        print(f"     {'TOTAL':.<30} {score}/100")

    # Strengths
    strengths = explanation.get("key_strengths", [])
    if strengths:
        print(f"\n  ✅ KEY STRENGTHS:")
        for s in strengths[:4]:
            if isinstance(s, dict):
                print(f"     • {s.get('strength', '')}")

    # Gaps
    gaps = explanation.get("key_gaps", [])
    if gaps:
        print(f"\n  ❌ KEY GAPS:")
        for g in gaps[:4]:
            if isinstance(g, dict):
                print(f"     • {g.get('gap', '')}")

    # Recommendation
    rec = explanation.get("final_recommendation", {})
    if isinstance(rec, dict):
        print(f"\n  🎯 RECOMMENDATION: {rec.get('decision', 'N/A')}")
        print(f"     Confidence: {rec.get('confidence', 'N/A')}")
        print(f"     Reason: {rec.get('reasoning', 'N/A')}")

    print(f"\n{'='*70}\n")


print("✅ Helper functions ready!")
````

````python
import time
from langsmith import traceable


@traceable(name="full_screening_pipeline", tags=["resume-screening"])
def screen_candidate(resume, job_description, candidate_label):
    """
    Run the complete 4-step screening pipeline for one candidate.

    Pipeline: Resume → Extract → Match → Score → Explain

    All steps are traced in LangSmith.
    """
    print(f"\n🚀 Screening: {candidate_label}")
    print(f"{'─'*50}")
    results = {}

    # STEP 1: Extract
    print("  📋 Step 1/4: Extracting skills...")
    results["extraction"] = extraction_chain.invoke({"resume": resume})
    print("     ✅ Done")
    time.sleep(10)  # Increased sleep to avoid rate limits

    # STEP 2: Match
    print("  🔍 Step 2/4: Matching against job requirements...")
    results["matching"] = matching_chain.invoke({
        "extracted_data": results["extraction"],
        "job_description": job_description,
    })
    print("     ✅ Done")
    time.sleep(10) # Increased sleep

    # STEP 3: Score
    print("  📊 Step 3/4: Scoring candidate...")
    results["scoring"] = scoring_chain.invoke({
        "matching_result": results["matching"],
        "extracted_data": results["extraction"],
        "job_description": job_description,
    })
    print("     ✅ Done")
    time.sleep(10) # Increased sleep

    # STEP 4: Explain
    print("  💡 Step 4/4: Generating explanation...")
    results["explanation"] = explanation_chain.invoke({
        "extracted_data": results["extraction"],
        "matching_result": results["matching"],
        "scoring_result": results["scoring"],
        "job_description": job_description,
    })
    print("     ✅ Done")

    return results


print("✅ Pipeline function ready!")
````

````python
# ============================================
# 🎯 RUN SCREENING FOR ALL 3 CANDIDATES
# ============================================

import time

print("🤖" * 35)
print("  AI RESUME SCREENING SYSTEM")
print("  Powered by LangChain + Google Gemini (FREE)")
print("  Traced by LangSmith (FREE)")
print("🤖" * 35)

all_results = {}

for key, data in RESUMES.items():
    label = f"{data['label']} - {data['name']}"

    # Run the pipeline
    results = screen_candidate(
        resume=data["resume"],
        job_description=JOB_DESCRIPTION,
        candidate_label=label,
    )

    all_results[key] = results

    # Print the report
    print_report(results, label)

    # Delay between candidates for rate limiting
    time.sleep(60) # Increased sleep to be more conservative for rate limits

print("\n✅ All 3 candidates screened successfully!")
````

````python
# ============================================
# 📊 CANDIDATE COMPARISON SUMMARY
# ============================================

print(f"\n{'#'*70}")
print(f"  📊 CANDIDATE COMPARISON SUMMARY")
print(f"{'#'*70}\n")

print(f"  {'Rank':<6} {'Name':<25} {'Type':<18} {'Score':<10} {'Decision'}")
print(f"  {'─'*6} {'─'*25} {'─'*18} {'─'*10} {'─'*25}")

ranking = []

for key, results in all_results.items():
    scoring = parse_json_safely(results.get("scoring", "{}"))
    explanation = parse_json_safely(results.get("explanation", "{}"))

    name = RESUMES[key]["name"]
    label = RESUMES[key]["label"]
    score = scoring.get("total_score", explanation.get("score", 0))
    rec = explanation.get("final_recommendation", {})
    decision = rec.get("decision", "N/A") if isinstance(rec, dict) else "N/A"

    ranking.append({"name": name, "label": label, "score": score, "decision": decision})

# Sort by score
ranking.sort(key=lambda x: x["score"] if isinstance(x["score"], (int, float)) else 0, reverse=True)

for i, r in enumerate(ranking, 1):
    emoji = "🥇" if i == 1 else "🥈" if i == 2 else "🥉"
    print(f"  {emoji} {i:<4} {r['name']:<25} {r['label']:<18} {str(r['score']):<10} {r['decision']}")

print(f"\n  📋 Job Position: Senior Data Scientist")
print(f"  🏢 Company: TechVision Analytics")
print(f"  📊 Total Candidates Screened: {len(ranking)}")
````

````python
# ============================================
# 🐛 DEBUGGING DEMO - Testing Edge Cases
# ============================================

@traceable(name="debug_incorrect_output", tags=["debugging", "edge-case"])
def debug_test():
    """
    Test with intentionally vague resume to verify
    the system doesn't hallucinate skills.

    This run will be visible in LangSmith for debugging.
    """
    print(f"\n{'#'*70}")
    print(f"  🐛 DEBUGGING DEMO: Testing with Vague/Ambiguous Resume")
    print(f"{'#'*70}\n")

    vague_resume = """
    Name: Test User
    I know some stuff about computers.
    I did some work before. I went to school.
    I have skills and I'm a hard worker.
    """

    print(f"  Input Resume: '{vague_resume.strip()}'")
    print(f"\n  🔍 Testing if system halluccinates skills...\n")

    # Run extraction on vague resume
    result = extraction_chain.invoke({"resume": vague_resume})
    parsed = parse_json_safely(result)

    print(f"  📊 Extraction Result:")
    print(json.dumps(parsed, indent=4))

    # Check for hallucination
    skills = parsed.get("technical_skills", [])
    ml_frameworks = parsed.get("ml_frameworks", [])
    languages = parsed.get("programming_languages", [])

    print(f"\n  🔍 HALLUCINATION CHECK:")
    print(f"     Technical skills found: {len(skills)} → {skills}")
    print(f"     ML frameworks found: {len(ml_frameworks)} → {ml_frameworks}")
    print(f"     Languages found: {len(languages)} → {languages}")

    if skills or ml_frameworks or languages:
        print(f"\n  ⚠️  WARNING: System may have hallucinated skills!")
        print(f"     The vague resume doesn't mention specific skills.")
        print(f"     → Check this run in LangSmith for debugging.")
    else:
        print(f"\n  ✅ PASSED: System correctly found no specific skills.")
        print(f"     Anti-hallucination guardrails working properly!")

    return parsed


debug_result = debug_test()
````
